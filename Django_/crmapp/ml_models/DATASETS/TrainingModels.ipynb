The dataset file 'dataset3_all_svc.csv' was obtained by extracting the features (through the functions available in Descriptor module of Propythia) from the original dataset ('dataset3_usar') where the non-positive results are composed with both transmembrane domains sequences and non-FPep and non-transmembrane domains sequences. Those features went through a feature reduction process using an SVC model, which, according to Sequeira's work, is the feature selection process that tends to produce better ML models for this dataset

```python
import pandas as pd
from propythia.machine_learning import MachineLearning
import pickle

def training(model):
    print(model)
    if model == 'svm':
        filename = 'crmapp/ml_models/dataset3_all_' + model+ '.sav'
    else:
        filename = 'crmapp/ml_models/dataset3_all_' + model+ '_model.sav'
    dataset_in = r'dataset3_all_svc.csv'
    dataset = pd.read_csv(dataset_in, delimiter=',')
    x_original = dataset.loc[:, dataset.columns != 'labels']
    labels = dataset['labels']
    # create Machine learning object
    ml = MachineLearning(x_original, labels, classes=['non_vfp', 'vfp'])
    best_model = ml.train_best_model(model)
    print(ml.score_testset(best_model))
    # save the model to disk
    # pickle.dump(best_model, open(filename, 'wb'))

for i in ['svm', 'rf',
          'sgd', 'gboosting',
          'lr', 'nn',
          'gnb', 'knn']:
    training(i)
```

The Propythia's "train_best_model" functions already provides all the necessary hyperparameter optimization and cross validation methodologies to obtain better models and to anlyze theis capabilities of prediction.
